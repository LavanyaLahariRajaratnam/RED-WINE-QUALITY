## <font color=blue>4. Modelling:</font>

### <font color=blue> a. Split the dataset using “train-test-split” function.</font>

### <font color=blue> b. Apply KNN classification on “quality” column of the dataset. Select the appropriate features
</font>

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier 

y=df['quality'].values
x=df.drop(['quality'],axis=1).values
```

```python
df.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=45,stratify=y)

knn=KNeighborsClassifier(n_neighbors=11)

knn.fit(x_train,y_train)

y_pred=knn.predict(x_test)
 
```

### <font color=blue>  c. Predict on the test set.</font>

```python
y_pred
```

### <font color=blue>d. Find out the accuracy.</font>

```python
from sklearn.metrics import accuracy_score
```

```python
print("accuracy: {}".format(accuracy_score(y_test, y_pred)))
```

```python
print('percentage Accuracy using KNN is : ',100*accuracy_score(y_test,y_pred))
```
